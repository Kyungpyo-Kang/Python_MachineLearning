# OpenCV
### 오픈소스 이미지(동영상) 라이브러리
이미지 형식 변환, 필터 처리같은 간단한 처리부터 얼굴인식, 물체인식, 문자인식 등 이미지와 관련된 다양한 처리를 할 수 있다. <br/>
이미지를 머신러닝의 입력으로 전달하려면 이미지를 숫자 배열 데이터로 변경해야 한다. <br/>
이미지는 일반적으로 BMP, PNG, JPEG 등 다양한 형태로 존재한다. <br/>
OpenCV는 이러한 이미지의 형식을 가공하고 색상을 맞출 때 사용한다.

```python
import urllib.request as req
url = "https://cdn2.iconfinder.com/data/icons/weather-color-2/500/weather-03-256.png"
req.urlretrieve(url, "test.png")
```

```python
import cv2
# img = cv2.imread("test1.png")
# 데이터가 없을 경우 None이 출력됨

img = cv2.imread("test.png")
# print(img)
```

```python
import matplotlib.pyplot as plt
import cv2
img = cv2.imread('test.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# imread()로 읽어들인 함수를 그대로 imshow()로 전달한 것이 아니라 cvtColor()함수를 이용해 색공간을
# BGR에서 RGB로 변환한다.
# OpenVC의 컬러 데이터는 BGR(파란색, 녹색, 빨간색) 순서로 되어있고, matplotlib의 컬러 데이터는
# RGB 순서로 되어있기 때문이다.
plt.show()
```

```python
import cv2


# OpenCV는 BMP/PPM/PGM/PBM/JPEG/JPEG2000/PNG/TIFF/OPENEXR/WebP 등 대표적인 이미지 형식을
# 모두 지원한다.
img = cv2.imread("test.png")
cv2.imwrite("out.jpg",img)
```

```python
img = cv2.imread('test.png')
img2 = cv2.resize(img, (300, 200))
cv2.imwrite('out-resize.png', img2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
img = cv2.imread('test.png')
img2 = img[75:225, 75:225]
# 이미지를 자를 때는 리스트 슬라이스를 사용해 배열[y1:y2, x1:x2] 형식으로 자른다.
cv2.imwrite("cut-resize.png", img2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
