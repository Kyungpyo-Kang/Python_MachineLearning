# 머신러닝이란?
인간의 학습 능력과 같은 기능을 컴퓨터가 하게 만드는 기술. <br/>
인공지능 연구 분야의 하나이며, 지적 능력을 컴퓨터가 하게 만드는 기술.
### <머신러닝으로 할 수 있는 것>
+ 분류: 주어진 데이터 분류하기
+ 회귀: 과거의 수치를 기반으로 미래의 수치 예측하기
+ 클러스터링: 데이터를 비슷한 집합으로 분류하기
+ 추천: 관련된 데이터 제공하기
+ 차원 축소: 데이터의 실질적인 특징과 관련없는 특징 없애기

### <머신러닝의 종류>
#### 교사 학습(지도 학습)
+ 데이터와 함께 정답이 주어지는 학습법
+ 미지의 데이터에서 예측할 수 있는 학습법
#### 비교사 학습(비지도 학습)
+ 정답 데이터가 주어지지 않은 학습법
+ 미지의 데이터에서 규칙성을 발견하는 학습법
#### 강화 학습
+ 행동을 기반으로 정답인지 알려주는 학습법
+ 데이터에서 최적의 답을 찾아내는 학습법

### <데이터를 수집하는 방법>
+ SNS와 블로그를 통한 데이터 수집
+ 쇼핑 상품을 통한 데이터 수집
+ 금융정보를 통해 데이터 수집
+ 오픈 데이터를 통한 데이터 수집
+ 이미지, 동영상, 음성데이터 수집
+ 사전 데이터 수집
+ 머신러닝 전용 데이터 수집
